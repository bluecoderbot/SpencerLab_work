```python
def getNeighbors_CO(arr):
    
    arr2 = (arr.copy())*0
    
    
    #IN YZ PLANE
    
    #bottom front
    arr2[:-1,:-1] += arr[1:,1:]
    mask = (arr2==0)
    #bottom 
    arr2[:-1] += arr[1:]*mask[:-1]
    mask = (arr2==0)
    #bottom back
    arr2[:-1, 1:] += arr[1:, :-1]*mask[:-1,1:]
    mask = (arr2==0)
    #top front
    arr2[1:, :-1] += arr[:-1, 1:]*mask[1:, :-1]
    mask = (arr2==0)
    #top
    arr2[1:] += arr[:-1]*mask[1:]
    mask = (arr2==0)
    
    #top back
    arr2[1:, 1:] += arr[:-1, :-1]*mask[1:, 1:]
    mask = (arr2==0)
    
    #front
    arr2[:, :-1] += arr[:, 1:]*mask[:, :-1]
    mask = (arr2==0)
    
    #back
    arr2[:, 1:] += arr[:, :-1]*mask[:, 1:]
    mask = (arr2==0)
    
    
    
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT 
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT
    
    #bottom front
    arr2[:-1,:-1, 1:] += arr[1:,1:, :-1]*mask[:-1,:-1, 1:]
    mask = (arr2==0)
    #bottom 
    arr2[:-1, :, 1:] += arr[1:, :, :-1]*mask[:-1, :, 1:]
    mask = (arr2==0)
    #bottom back
    arr2[:-1, 1:, 1:] += arr[1:, :-1, :-1]*mask[:-1, 1:, 1:]
    mask = (arr2==0)
    
    #top front
    arr2[1:, :-1, 1:] += arr[:-1, 1:, :-1]*mask[1:, :-1, 1:]
    mask = (arr2==0)
    #top
    arr2[1:, :, 1:] += arr[:-1, :, :-1]*mask[1:, :, 1:]
    mask = (arr2==0)
    #top back
    arr2[1:, 1:, 1:] += arr[:-1, :-1, :-1]*mask[1:, 1:, 1:]
    mask = (arr2==0)
    
    #front
    arr2[:, :-1, 1:] += arr[:, 1:, :-1]*mask[:, :-1, 1:]
    mask = (arr2==0)
    #back
    arr2[:, 1:, 1:] += arr[:, :-1, :-1]*mask[:, 1:, 1:]
    mask = (arr2==0)
    
    #just left 
    arr2[:, :, 1:] += arr[:, :, :-1]*mask[:, :, 1:]
    mask = (arr2==0)
    
    
    
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    
    #bottom front
    arr2[:-1,:-1, :-1] += arr[1:,1:, 1:]*mask[:-1,:-1, :-1]
    mask = (arr2==0)
    #bottom 
    arr2[:-1, :, :-1] += arr[1:, :, 1:]*mask[:-1, :, :-1]
    mask = (arr2==0)
    #bottom back
    arr2[:-1, 1:, :-1] += arr[1:, :-1, 1:]*mask[:-1, 1:, :-1]
    mask = (arr2==0)
    
    #top front
    arr2[1:, :-1, :-1] += arr[:-1, 1:, 1:]*mask[1:, :-1, :-1]
    mask = (arr2==0)
    #top
    arr2[1:, :, :-1] += arr[:-1, :, 1:]*mask[1:, :, :-1] 
    mask = (arr2==0)
    #top back
    arr2[1:, 1:, :-1] += arr[:-1, :-1, 1:]*mask[1:, 1:, :-1]
    mask = (arr2==0)
    
    #front
    arr2[:, :-1, :-1] += arr[:, 1:, 1:]*mask[:, :-1, :-1]
    mask = (arr2==0)
    #back
    arr2[:, 1:, :-1] += arr[:, :-1, 1:]*mask[:, 1:, :-1]
    mask = (arr2==0)
    
    
    #just right 
    arr2[:, :, :-1] += arr[:, :, 1:]*mask[:, :, :-1]
    
    
    
    return arr2*(arr!=0)
```

```python
def getNeighbors(arr):
    
    arr2 = (arr.copy())*0
    
    
    #IN YZ PLANE
    
    #bottom front
    arr2[:-1,:-1] += arr[1:,1:]
    
    
    
    #bottom 
    arr2[:-1] += arr[1:]
    
    
    #bottom back
    arr2[:-1, 1:] += arr[1:, :-1]
    
    #top front
    arr2[1:, :-1] += arr[:-1, 1:]
    #top
    arr2[1:] += arr[:-1]
    #top back
    arr2[1:, 1:] += arr[:-1, :-1]
    
    #front
    arr2[:, :-1] += arr[:, 1:]
    #back
    arr2[:, 1:] += arr[:, :-1]
    
    
    
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT 
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT
    #LEFT LEFT LEFT LEFT LEFT LEFT LEFT
    
    #bottom front
    arr2[:-1,:-1, 1:] += arr[1:,1:, :-1]
    #bottom 
    arr2[:-1, :, 1:] += arr[1:, :, :-1]
    #bottom back
    arr2[:-1, 1:, 1:] += arr[1:, :-1, :-1]
    
    #top front
    arr2[1:, :-1, 1:] += arr[:-1, 1:, :-1]
    #top
    arr2[1:, :, 1:] += arr[:-1, :, :-1]
    #top back
    arr2[1:, 1:, 1:] += arr[:-1, :-1, :-1]
    
    #front
    arr2[:, :-1, 1:] += arr[:, 1:, :-1]
    #back
    arr2[:, 1:, 1:] += arr[:, :-1, :-1]
    
    #just left 
    arr2[:, :, 1:] += arr[:, :, :-1]
    
    
    
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    #RIGHT RIGHT RIGHT RIGHT RIGHT RIGHT
    
    #bottom front
    arr2[:-1,:-1, :-1] += arr[1:,1:, 1:]
    #bottom 
    arr2[:-1, :, :-1] += arr[1:, :, 1:]
    #bottom back
    arr2[:-1, 1:, :-1] += arr[1:, :-1, 1:]
    
    #top front
    arr2[1:, :-1, :-1] += arr[:-1, 1:, 1:]
    #top
    arr2[1:, :, :-1] += arr[:-1, :, 1:]
    #top back
    arr2[1:, 1:, :-1] += arr[:-1, :-1, 1:]
    
    #front
    arr2[:, :-1, :-1] += arr[:, 1:, 1:]
    #back
    arr2[:, 1:, :-1] += arr[:, :-1, 1:]
    
    
    #just right 
    arr2[:, :, :-1] += arr[:, :, 1:]
    
    
    
    return arr2*(arr!=0)
```

```python
def get_centroids(arr):
    
    arr= arr!=0
    markers, nummarkers = ndi.label(arr) 
    
    points = np.where(arr)
    points = np.asarray(points)
    
    pt_labels = markers[points[0], points[1], points[2]]
    
    Z_mass = (np.bincount(pt_labels, weights=points[0]))[1:]

    Y_mass = (np.bincount(pt_labels, weights=points[1]))[1:]

    X_mass = (np.bincount(pt_labels, weights=points[2]))[1:]

    label_occurances = (np.bincount(pt_labels, weights=np.ones((points.shape[1]))) )[1:]
    
    
    Z_centroid = Z_mass/label_occurances
    Y_centroid = Y_mass/label_occurances
    X_centroid = X_mass/label_occurances
    
    return Z_centroid, Y_centroid, X_centroid, label_occurances
```

```python
def get_occurances(arr):
    
    arr= arr!=0
    markers, nummarkers = ndi.label(arr) 
    
    points = np.where(arr)
    points = np.asarray(points)
    
    pt_labels = markers[points[0], points[1], points[2]]

    label_occurances = (np.bincount(pt_labels, weights=np.ones((points.shape[1]))) )[1:]
    
    
    return label_occurances
```

```python
def get_endpoints_occurrances(arr, epts):
    arr= arr!=0
    markers, nummarkers = ndi.label(arr) 
    
    arr = arr*(epts!=0)
    
    points = np.where(arr)
    points = np.asarray(points)
    
    pt_labels = markers[points[0], points[1], points[2]]

    label_occurances = (np.bincount(pt_labels, weights=np.ones((points.shape[1]))) )[1:]
    
    
    return label_occurances
    
```

```python
def pair_labels_W_vals(arr, labels, vals):
    arr= arr!=0
    markers, nummarkers = ndi.label(arr) 
    
    points = np.where(arr)
    points = np.asarray(points)
    
    pt_labels = markers[points[0], points[1], points[2]]
    
    dict_1 = dict(zip(labels, vals))
    
    new_labels = np.vectorize(dict_1.get)(pt_labels)
    
    markers[points[0], points[1], points[2]] = new_labels
    
    return markers
```
