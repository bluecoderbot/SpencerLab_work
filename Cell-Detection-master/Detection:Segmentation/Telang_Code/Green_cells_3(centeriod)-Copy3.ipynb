```python
import threading 
cells_centroid3 = []

def Centroid(num1,num2): 
    #for i in range(num1, len(unk)):
    for i in range(num1, num2):
        
        print(str(i)+"   ")
        co_ordinates =np.where(lab==unk[i])
        z_c = sum(co_ordinates[0])/len(co_ordinates[0])
        y_c = sum(co_ordinates[1])/len(co_ordinates[0])
        x_c = sum(co_ordinates[2])/len(co_ordinates[0])

        centroid = [z_c+1, y_c+1, x_c+1, len(co_ordinates[0])]
        cells_centroid3.append(centroid)

  
if __name__ == "__main__": 
    # creating thread 
    #x = len(unk)
    x = 20
    b = x/5
    b = int(b)
    t1 = threading.Thread(target=Centroid, args=(0,b+1,)) 
    t2 = threading.Thread(target=Centroid, args=(b,2*b+1,)) 
    t3 = threading.Thread(target=Centroid, args=(2*b,3*b+1,)) 
    t4 = threading.Thread(target=Centroid, args=(3*b,4*b+1,))
    t5 = threading.Thread(target=Centroid, args=(4*b,b+1,)) 
    # starting thread 1 
    t1.start() 
    t2.start() 
    t3.start() 
    t4.start()
    t5.start()

    # wait until thread 1 is completely executed 
    t1.join() 
    t2.join() 
    t3.join() 
    t4.join() 
    t5.join() 
    # both threads completely executed 
    print("Done!") 
    
    
    
    
print(cells_centroid3)
```
