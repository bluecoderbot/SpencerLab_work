```python
data = np.asarray(csv.values)
data[0,2]
```

```python
cellsnum= []
for i in range(len(data)):
    f= np.count_nonzero(data[:,14] == data[i,14])
    cellsnum.append(f)
```

```python
avsize=[]
clustermass=[]
avdistb=[]
avdistt=[]
for i in range(len(data)):

    total= 0
    dbc= 0
    dbt=0
    for d in range(len(data)):
        #cell size
        if data[d,14]==data[i,14]:
            total = total+int(data[d,4]) 
        #blood vessel
            dbc = dbc + float(data[d,9])
            dbt = dbt + float(data[d,13])
    
    
    app = total/cellsnum[i]
    avsize.append(app)
    clustermass.append(total)
    
    app2= dbc/cellsnum[i]
    avdistb.append(app2)
     
    app3= dbt/cellsnum[i]
    avdistt.append(app3)
    
```

```python

wx=[]
wy=[]
wz=[]
ux=[]
uy=[]
uz=[]
for i in range(len(data)):
    x= 0
    y=0
    z=0
    dist=0
    x1=0
    y1=0
    z1=0
    for d in range(len(data)):
        if data[d,14]==data[i,14]:
            x+= int(data[d,3]*data[d,4])
            y+=int(data[d,2]*data[d,4]) 
            z+=int(data[d,1]*data[d,4])
            x1 += int(data[d,3])
            y1 +=int(data[d,2] )
            z1 +=int(data[d,1])

    r=cellsnum[i]
    l=avsize[i]
    wx.append(x/(r*l))
    wy.append(y/(r*l))
    wz.append(z/(r*l))
    ux.append(x1/(r))
    uy.append(y1/(r))
    uz.append(z1/(r))
    
```

```python
uy
```

```python
wr=[]
ur=[]
for i in range(len(data)):
    dx=0
    dy=0
    dz=0
    az=0
    ay=0
    ax=0
    wdist=0
    udist=0
    for d in range(len(data)):
        if data[d,14]==data[i,14]:
            dz+= float(data[d,3]-wx[d])
            dy+= float(data[d,2]-wy[d])
            dz+= float(data[d,1]-wz[d])
            az+= float(data[d,3]-ux[d])
            ay+= float(data[d,2]-uy[d])
            az+= float(data[d,1]-uz[d])
            wdist+= (dx**2+dy**2+dz**2)**.5
            udist+= (ax**2+ay**2+az**2)**.5
    wr.append(wdist/cellsnum[i])
    ur.append(udist/cellsnum[i])
```

```python
ur
```

```python
god= np.column_stack((data[:,14],wz,wy,wx,uz,uy,ux,avsize,clustermass,wr,ur,avdistb,avdistt,cellsnum))
```
